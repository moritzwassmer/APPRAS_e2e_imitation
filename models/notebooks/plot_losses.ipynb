```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#path_lidar_stats = "/Users/julianvonklitzing/Documents/GitHub/end2endappras/models/notebooks/experiment_files/2023_02_26_10_43/stats/stats_performance.csv"
path_lidar_stats = "/Users/julianvonklitzing/Documents/GitHub/end2endappras/models/notebooks/experiment_files/2023_03_14_12_48_27/stats/stats_performance.csv"
df_lidar_stats = pd.read_csv(path_lidar_stats)
df_lidar_stats.index = list(range(1, len(df_lidar_stats)+1))
df_lidar_stats["train_loss"] = df_lidar_stats[df_lidar_stats.columns[1:4]].sum(axis=1)
df_lidar_stats["val_loss"] = df_lidar_stats[df_lidar_stats.columns[4:]].sum(axis=1)
```

```python
ax = df_lidar_stats[["train_loss", "val_loss"]].plot(xlabel="Epoch", ylabel="Total Loss", xticks=list(range(1, len(df_lidar_stats)+1)),title="Camera+LiDAR Model: Train and Validation Loss", grid=True) # 
fig = plt.gcf()
fig.savefig("camera_lidar_results.pdf")
```
